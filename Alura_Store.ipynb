```python
import matplotlib.pyplot as plt
import numpy as np

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificacion = [3.98, 4.04, 4.05, 4.00]
envio = [26018.61, 25216.24, 24805.68, 23459.46]
ingreso = [1150880400.0, 1116343500.0, 1098019600.0, 1038375700.0]
ventas = [2359, 2359, 2359, 2358]
producto = [60, 65, 57, 62]

def normalizar(valores):
    min_v, max_v = min(valores), max(valores)
    return [(v - min_v) / (max_v - min_v) * 10 for v in valores]

plt.figure(figsize=(12, 15))
plt.style.use('seaborn-v0_8-whitegrid')

plt.subplot(5, 1, 1)
plt.plot(tiendas, calificacion, marker='o', color='gold', linewidth=2.5)
plt.title('Calificación de las Tiendas', fontsize=13, fontweight='bold')
plt.ylabel('Calificación')
for i, v in enumerate(calificacion):
    plt.text(i, v + 0.004, f'{v}', ha='center', fontsize=10, color='#444')


plt.subplot(5, 1, 2)
plt.plot(tiendas, envio, marker='o', color='sandybrown', linewidth=2.5)
plt.title('Costo Promedio de Envío ($)', fontsize=13, fontweight='bold')
for i, v in enumerate(envio):
    plt.text(i, v + 200, f'${v:,.2f}', ha='center', fontsize=9, color='#444')

plt.subplot(5, 1, 3)
plt.plot(tiendas, ingreso, marker='o', color='cyan', linewidth=2.5)
plt.title('Ingreso Total de las Tiendas', fontsize=13, fontweight='bold')
for i, v in enumerate(ingreso):
    plt.text(i, v + 10000000, f'${v:,.2f}', ha='center', fontsize=9, color='#444')


plt.subplot(5, 1, 4)
plt.plot(tiendas, ventas, marker='o', color='darkmagenta', linewidth=2.5)
plt.title('Ventas Totales por Tienda', fontsize=13, fontweight='bold')
for i, v in enumerate(ventas):
    plt.text(i, v + 0.1, str(v), ha='center', fontsize=9, color='#444')

plt.subplot(5, 1, 5)
plt.plot(tiendas, producto, marker='o', color='red', linewidth=2.5)
plt.title('Cantidad del Producto Más Vendido', fontsize=13, fontweight='bold')
for i, v in enumerate(producto):
    plt.text(i, v + 0.3, str(v), ha='center', fontsize=9, color='#444')

plt.tight_layout()
plt.show()


puntajes = []
for i in range(len(tiendas)):
    score = calificacion[i] * 0.3 + (ventas[i] / max(ventas)) * 0.2 + (producto[i] / max(producto)) * 0.2 + (ingreso[i] / max(ingreso)) * 0.3
    puntajes.append(score)

menor_tienda = tiendas[np.argmin(puntajes)]

print(f"\n La tienda con menor desempeño general es: {menor_tienda}")
```
